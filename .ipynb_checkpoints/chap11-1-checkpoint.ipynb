```python
#サンプルデータ2
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np


X_list = []
y_list = []
n = np.arange(1.0, 2.0+0.1, 0.1)

for i,d in enumerate(n):
    X,y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=d, shuffle=True, random_state=0)
    X_list.append(X)
    y_list.append(y)

fig, ax = plt.subplots(nrows=len(n),  figsize=(5,3*len(n)))

for i,d in enumerate(n):
    ax[i].scatter(X_list[i][:,0], X_list[i][:,1], c='white', marker='o', edgecolor='black', s=50, label='std: %.1f'%(d))
    ax[i].legend(loc='lower left')
    ax[i].grid()

fig.tight_layout()

```

```python
#k-means その1
from sklearn.cluster import KMeans

km =KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)

plt.scatter(X[y_km==0, 0], X[y_km==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
plt.scatter(X[y_km==1, 0], X[y_km==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
plt.scatter(X[y_km==2, 0], X[y_km==2, 1], s=50, c='red', edgecolor='black', marker='v', label='cluster 3')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='*', c='red', edgecolor='black', label='cetroids')

plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#k-means その2
from sklearn.cluster import KMeans

for i,d in enumerate(n):
    km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    #km = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    y_km = km.fit_predict(X_list[i])
    
    plt.figure(figsize=(4,3))
    plt.scatter(X_list[i][y_km==0, 0], X_list[i][y_km==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
    plt.scatter(X_list[i][y_km==1, 0], X_list[i][y_km==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
    plt.scatter(X_list[i][y_km==2, 0], X_list[i][y_km==2, 1], s=50, c='red', edgecolor='black', marker='v', label='cluster 3')
    plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='*', c='red', edgecolor='black', label='cetroids')
    plt.legend(loc='lower left')
    plt.title("std: %.1f" % (d))
    plt.grid()
    plt.tight_layout()
```

```python
#k-means++
from sklearn.cluster import KMeans

fig, ax = plt.subplots(ncols=2, nrows=len(n),  figsize=(10,3*len(n)))

for i,d in enumerate(n):
    km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    y_km = km.fit_predict(X_list[i])
    
    ax[i,0].scatter(X_list[i][y_km==0, 0], X_list[i][y_km==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
    ax[i,0].scatter(X_list[i][y_km==1, 0], X_list[i][y_km==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
    ax[i,0].scatter(X_list[i][y_km==2, 0], X_list[i][y_km==2, 1], s=50, c='red', edgecolor='black', marker='v', label='cluster 3')
    ax[i,0].scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='*', c='red', edgecolor='black', label='cetroids')
    ax[i,0].legend(loc='lower left')
    ax[i,0].set_title("std: %.1f" % (d))
    ax[i,0].grid()

for i,d in enumerate(n):
    km = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    y_km = km.fit_predict(X_list[i])
    
    ax[i,1].scatter(X_list[i][y_km==0, 0], X_list[i][y_km==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
    ax[i,1].scatter(X_list[i][y_km==1, 0], X_list[i][y_km==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
    ax[i,1].scatter(X_list[i][y_km==2, 0], X_list[i][y_km==2, 1], s=50, c='red', edgecolor='black', marker='v', label='cluster 3')
    ax[i,1].scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='*', c='red', edgecolor='black', label='cetroids')
    ax[i,1].legend(loc='lower left')
    ax[i,1].set_title("k-means++ std: %.1f" % (d))
    ax[i,1].grid()    
    
fig.tight_layout()
```

```python
#エルボー法
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X,y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)

distortions = []

for i in range(1,11):
    km =KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('number of clusters')
plt.ylabel('Distortion')
plt.xticks(range(1,11))
plt.tight_layout()
plt.show()
```

```python
#シルエット図1

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples

X,y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)

km =KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []

for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper +=len(c_silhouette_vals)
    color = cm.jet(float(i)/n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=0.5, edgecolor='none', color=color)
    ytics.append((y_ax_upper+y_ax_lower)/2.0)
    y_ax_lower += len(c_silhouette_vals)

silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', linestyle='--')
plt.yticks(yticks, cluster_labels +1)
plt.ylabel('cluster')
plt.xlabel('silhouette coefficient')
plt.tight_layout()
plt.show()
```

```python
#シルエット図2 上手くいかない場合

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples

X,y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)

km =KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)

plt.figure(figsize=(6, 4))

plt.scatter(X[y_km==0, 0], X[y_km==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
plt.scatter(X[y_km==1, 0], X[y_km==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
plt.scatter(X[y_km==2, 0], X[y_km==2, 1], s=50, c='red', edgecolor='black', marker='v', label='cluster 3')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='*', c='red', edgecolor='black', label='cetroids')
plt.grid()
plt.tight_layout()

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []

plt.figure(figsize=(7, 5))

for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper +=len(c_silhouette_vals)
    color = cm.jet(float(i)/n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=0.5, edgecolor='none', color=color)
    ytics.append((y_ax_upper+y_ax_lower)/2.0)
    y_ax_lower += len(c_silhouette_vals)

silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', linestyle='--')
plt.yticks(yticks, cluster_labels +1)
plt.ylabel('cluster')
plt.xlabel('silhouette coefficient')
plt.tight_layout()
plt.show()
```
