```python
#housingデータ取得
import pandas as pd

df = pd.read_csv('./python-machine-learning-book-2nd-edition-master/code/ch10/housing.data.txt', header=None, sep='\s+')

df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 
              'NOX', 'RM', 'AGE', 'DIS', 'RAD', 
              'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df
```

```python
#線形回帰モデルの性能評価
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RANSACRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import Ridge, Lasso


X = df.iloc[:, :-1].values
y = df['MEDV'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

slr1 = LinearRegression()
slr2 = Ridge(alpha=1.0)
slr3 = Lasso(alpha=1.0)


slrs = [slr1, slr2, slr3]
slr_names =['LinearRegression', 'Ridge(r^2)', 'Lasso(r^1)']
results_ = []

for slr, slr_name in zip(slrs,slr_names):
    print(slr_name)
    slr.fit(X_train, y_train)
    y_train_pred = slr.predict(X_train)
    y_test_pred = slr.predict(X_test)

    list_temp = []
    
#     print(np.abs(slr.coef_))
    
    for i,coef in enumerate(slr.coef_):
#         list_temp.append(coef)
        list_temp.append(coef/np.abs(slr.coef_).sum())
#         print('coef[%d]=%.1f' % (i, coef))
        
    
    
#     print('MSE train: %.3f, test: %.3f' % (mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred)))
#     print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred)))
#     print('')
    
    list_temp.append(mean_squared_error(y_train, y_train_pred))
    list_temp.append(mean_squared_error(y_test, y_test_pred))
    list_temp.append(r2_score(y_train, y_train_pred))
    list_temp.append(r2_score(y_test, y_test_pred))
    
    results_.append(list_temp)
    
results = pd.DataFrame(results_)
results.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX',
                   'RM', 'AGE', 'DIS', 'RAD', 'TAX',
                   'PTRATIO', 'B', 'LSTAT',
                   'MSE train', 'MSE test', 'R^2 train', 'R^2 test']
print(results.T)

plt.scatter(y_train_pred, (y_train_pred - y_train), c='steelblue', edgecolor='white', marker='o', label='Training data')
plt.scatter(y_test_pred, (y_test_pred - y_test), c='limegreen', edgecolor='white', marker='o', label='Test data')

plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, color='black', lw=2)
plt.xlim([-10, 50])
plt.tight_layout()

# plt.savefig('images/10_09.png', dpi=300)
plt.show()
```

```python
#線形回帰モデルの性能評価
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RANSACRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np



X = df.iloc[:, :-1].values
y = df['MEDV'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

slr1 = RANSACRegressor(LinearRegression(), max_trials=100, min_samples=50, loss='absolute_loss', residual_threshold=5.0, random_state=0)
slr2 = RANSACRegressor(Ridge(alpha=1.0), max_trials=100, min_samples=50, loss='absolute_loss', residual_threshold=5.0, random_state=0)
slr3 = RANSACRegressor(Lasso(alpha=1.0), max_trials=100, min_samples=50, loss='absolute_loss', residual_threshold=5.0, random_state=0)


slrs = [slr1, slr2, slr3]
slr_names =['LinearRegression(RANSAC)', 'Ridge:r^2(RANSAC)', 'Lasso:r^1(RANSAC)']
results_ = []

for slr, slr_name in zip(slrs,slr_names):
    print(slr_name)
    slr.fit(X_train, y_train)
    y_train_pred = slr.predict(X_train)
    y_test_pred = slr.predict(X_test)

    list_temp = []
    
#     print(np.abs(slr.coef_))
    
    for i,coef in enumerate(slr.estimator_.coef_):
#         list_temp.append(coef)
        list_temp.append(coef/np.abs(slr.estimator_.coef_).sum())
#         print('coef[%d]=%.1f' % (i, coef))
        
    
    
#     print('MSE train: %.3f, test: %.3f' % (mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred)))
#     print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred)))
#     print('')
    
    list_temp.append(mean_squared_error(y_train, y_train_pred))
    list_temp.append(mean_squared_error(y_test, y_test_pred))
    list_temp.append(r2_score(y_train, y_train_pred))
    list_temp.append(r2_score(y_test, y_test_pred))
    
    results_.append(list_temp)
    
results = pd.DataFrame(results_)
results.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX',
                   'RM', 'AGE', 'DIS', 'RAD', 'TAX',
                   'PTRATIO', 'B', 'LSTAT',
                   'MSE train', 'MSE test', 'R^2 train', 'R^2 test']
print(results.T)

plt.scatter(y_train_pred, (y_train_pred - y_train), c='steelblue', edgecolor='white', marker='o', label='Training data')
plt.scatter(y_test_pred, (y_test_pred - y_test), c='limegreen', edgecolor='white', marker='o', label='Test data')

plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, color='black', lw=2)
plt.xlim([-10, 50])
plt.tight_layout()

# plt.savefig('images/10_09.png', dpi=300)
plt.show()
```

```python
#データの視覚化1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import copy

df = pd.read_csv('./python-machine-learning-book-2nd-edition-master/code/ch10/housing.data.txt', header=None, sep='\s+')

df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 
              'NOX', 'RM', 'AGE', 'DIS', 'RAD', 
              'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

cols = ['NOX', 'RM', 'DIS', 'PTRATIO', 'LSTAT', 'B', 'MEDV']

df = df.loc[:, cols]
df2 = copy.deepcopy(df)
df2.iloc[:, 0] = -0.5*df.iloc[:,0] + 0.3*df.iloc[:,1]
df2.iloc[:, 1] = -0.2*df.iloc[:,0] + 0.5*df.iloc[:,1]
df2.iloc[:, 2] = -0.3*df.iloc[:,3] - 0.3*df.iloc[:,4]
df2.iloc[:, 3] = -0.6*df.iloc[:,0]
df2.iloc[:, 4] = -0.5*df.iloc[:,0] + 0.2*df.iloc[:,1]
df2.iloc[:, 5] =  0.2*df.iloc[:,1] - 0.2*df.iloc[:,2] - 0.2*df.iloc[:,3] - 0.2*df.iloc[:,4]

df2.columns = ['1', '2', '3', '4', '5', '6', 'MEDV']

df2.head()

sns.pairplot(df2, height=2.5)

plt.tight_layout()
plt.show()
```
