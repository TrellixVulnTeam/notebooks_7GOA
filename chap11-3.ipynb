```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=200, noise=0.1, random_state=0)

plt.scatter(X[:,0], X[:,1])
plt.tight_layout()
plt.show()
```

```python
#k-means, tree
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np

km =KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)

ac = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
y_ac= ac.fit_predict(X)
x_ac_centers_= []
y_ac_centers_= []
x_ac_centers_.append(np.mean(X[y_ac==0, 0]))
x_ac_centers_.append(np.mean(X[y_ac==1, 0]))
y_ac_centers_.append(np.mean(X[y_ac==0, 1]))
y_ac_centers_.append(np.mean(X[y_ac==1, 1]))

f, (ax1, ax2) = plt.subplots(1,2, figsize=(8,3))

ax1.scatter(X[y_km==0, 0], X[y_km==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
ax1.scatter(X[y_km==1, 0], X[y_km==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
ax1.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='*', c='red', edgecolor='black', label='cetroids')
ax1.set_title('K-means clustering')

ax2.scatter(X[y_ac==0, 0], X[y_ac==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
ax2.scatter(X[y_ac==1, 0], X[y_ac==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
ax2.scatter(x_ac_centers_, y_ac_centers_, s=250, marker='*', c='red', edgecolor='black', label='cetroids')
ax2.set_title('Agglomerative clustering')

plt.legend
plt.tight_layout()
plt.show()
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.2, min_samples=3, metric='euclidean')
y_db = db.fit_predict(X)
x_db_centers_= []
y_db_centers_= []
x_db_centers_.append(np.mean(X[y_db==0, 0]))
x_db_centers_.append(np.mean(X[y_db==1, 0]))
y_db_centers_.append(np.mean(X[y_db==0, 1]))
y_db_centers_.append(np.mean(X[y_db==1, 1]))

plt.scatter(X[y_db==0, 0], X[y_db==0, 1], s=50, c='lightgreen', edgecolor='black', marker='s', label='cluster 1')
plt.scatter(X[y_db==1, 0], X[y_db==1, 1], s=50, c='orange', edgecolor='black', marker='o', label='cluster 2')
# plt.scatter(x_ac_centers_, y_ac_centers_, s=250, marker='*', c='red', edgecolor='black', label='cetroids')
plt.title('DBSCAN clustering')

plt.legend
plt.tight_layout()
plt.show()
```
